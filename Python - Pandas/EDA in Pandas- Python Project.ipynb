```python
# correlation amoung numerica values in the dataset
df.corr(numeric_only = True)
```

```python
# visualization of correlation using heatmap
sns.heatmap(df.corr(numeric_only = True), annot = True)

# change the size of the heatmap
plt.rcParams['figure.figsize'] = (20, 7)
plt.show()
```

```python
# getting 2020 Population mean for all countires together in each contient
df.groupby('Continent')['2020 Population'].mean()
```

```python
# getting all data for a particular continent
df[df['Continent'].str.contains('Oceania')]
```

```python
# getting least populated country in each continent based on 2020 Population 
df.groupby('Continent').max().sort_values(by = '2022 Population', ascending = False)
```

```python
# plot chart for most populated country in each continent based on 2020 Population 
df2 = df.groupby('Continent').max().sort_values(by = '2022 Population', ascending = False)

df2.plot()
```

```python
# switch row and columns of previous dataset and select columns with numeric values to create a plot
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].max().sort_values(by = '2022 Population', ascending = False)
df3 = df2.transpose()
df3
```

```python
df3.plot(title = 'Highest population in each continent for last 50 years')
```
