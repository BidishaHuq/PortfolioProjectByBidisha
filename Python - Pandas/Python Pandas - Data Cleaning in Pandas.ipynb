```python
import pandas as pd
```

```python
# reading dataset

df = pd.read_excel(r"C:\Users\bhuq8\Downloads\Customer Call List.xlsx")

df
```

```python
# removing duplicate data 
df = df.drop_duplicates()
```

```python
# removing unnecessary column
df = df.drop(columns = 'Not_Useful_Column')
```

```python
# removing unwanted characters from Last_Name column
df['Last_Name'] =df['Last_Name'].str.strip("123/..._")
```

```python
# removing unwanted characters from Phone_Number column
df['Phone_Number'] =df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '' , regex = True)
```

```python
# converting the data type of Phone_Number column
df['Phone_Number'] = df['Phone_Number'].astype(str)
```

```python
# formatting the Phone Numbers
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3]+'-'+x[3:6]+ '-'+x[6:10])
```

```python
# validating data
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '', regex = True)

df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '', regex = True)
```

```python
# splitting address into three columns and assinging them into a new dataframe
df2 = df['Address'].str.split(',', expand = True)

df2
```

```python
# creating three new columns in the original dataframe and assigning values from the new dataframe
df[['Street', 'State', 'Postcode']] = df2
```

```python
# dropping column 'Address' as the work is done on it
df = df.drop(columns = 'Address')
```

```python
# converting 'Yes' into 'Y'  and 'No' to 'N' in Paying Customer column
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
```

```python
# converting 'Yes' into 'Y'  and 'No' to 'N' in Do_Not_Contact column
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
```

```python
# checking the dataframe to see if everything is alright
df
```

```python
# handling missing values
df= df.replace('N/a', '')
df= df.replace('NaN', '')
```

```python
# dropping the rows where customers are wished not to contacted 
for x in df.index:
    if df.loc[x, "Do_Not_Contact"]  == 'Y':
        df.drop(x, inplace = True)
```

```python
df
```

```python
# dropping the rows where Phone Number are missing
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace = True)
```

```python
# handling missing values
df = df.fillna('')
```

```python
# reseting the index
df.reset_index(drop = True)
```
