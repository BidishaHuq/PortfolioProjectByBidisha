# Data cleaning

```python
# importing pandas the dataset
import pandas as pd

df = pd.read_csv(r"C:\Users\bhuq8\OneDrive\Bootcamp think employment\python project\retail.csv")

df
```

```python
# checking for duplicates
df = df.drop_duplicates()
```

```python
# getting the statistical data
df.describe()
```

```python
# finding out if there is any missing value
df.info()
```

```python
# getting all values for country to deal with missing value
df['Country'].unique()
```

```python
# replacing missing value with 'United States'
df['Country'] = df['Country'].fillna('United States')
```

```python
# getting data from States
df['State'].unique()
```

```python
# implementating data validation
df['State'] = df['State'].str.title()
```

```python
# repeating same process for City
df['City'].unique()
```

```python
df['City'] = df['City'].str.title()
```

```python
# calculating the total sale in a new column
df['Total Sales'] = round(df['Sales']*df['Quantity'], 2)
```

```python
# calculating the total cost in a new column
df['Total Cost'] = round(df['Cost']*df['Quantity'], 2)
```

```python
# calculating the total profit in a new column
df['Total Profit'] = df['Total Sales'] - df['Total Cost']

# also calculating profit per unit
df['Unit Profit'] = df['Sales'] - df['Cost']
```

```python
df
```

```python
# converting date (object data type) to date time
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
```

```python
# repeating same step for Shipping Date
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
```

```python
# finding out the differences between order and shipping dates
df['Processing Days'] = df['Ship Date'] - df['Order Date']
```

```python
# slicing Order Dates into year /quarter/ month
df['Year'] = df['Order Date'].dt.year
df['Quarter'] = df['Order Date'].dt.quarter
df['MonthNo'] = df['Order Date'].dt.month
```

```python
# converting month no into month name
df['Month'] = df['MonthNo'].apply(lambda x: pd.Timestamp(year = 2000, month = x, day =1).strftime('%B'))
# or 
#df['Month'] = df['Order Date'].dt.month_name()
```

```python
# dropping column MonthNo
df = df.drop(columns='MonthNo')
```

```python
# converting int to string to fix Quarter column
qtr = "Qtr "
df['Quarter'] = qtr+ df['Quarter'].astype(str)
```

```python
# splitting customer's first and lastname and assinging into a new data frame
df2 = df['Customer Name'].str.split(' ', expand = True)
df2.head(5)
```

```python
df2[2].unique()
```

```python
# adding both part of last name together and drop middle two columns
df2[3] = df2[1] + ' ' + df2[2].fillna('')  
df2= df2.drop(columns = [1,2])
df2
```

```python
# assinging df2 into df's column
df[['First Name','Last Name']] = df2
```

```python
# dropping column Customer Name
df = df.drop(columns = 'Customer Name')
```

# Answering the given Questions

```python
# Q1.	Calculate the total sales from the dataset

print(f"The total sales from the data set is: {round(df['Total Sales'].sum(), 2)}")
```

```python
# Q2.	Determine the average quantity sold for each product category

# finding out how many categories is there
category = df['Category'].unique()

for key in category:
    df_key = df.loc[(df['Category'] == key)]
    print(f"The average quantity sold for {key} category {round(df_key['Quantity'].mean(), 2)}\n")
```

```python
# Q3.	Identify the most frequently used ship mode.

print(f"The most frequently used Ship mode is: {df['Ship Mode'].max()}")
```

```python
# Q4.	Find the top 5 customers based on their total sales.

# grouping by Customers' Id and Name then perform aggregration to get the summation of total sales
total_sales = df.groupby(['Customer ID', 'First Name', 'Last Name'])

df_total_sales = total_sales.agg({'Total Sales': 'sum'}).sort_values(by = 'Total Sales', ascending = False).head(5)

print("The top 5 customers based on their total sales are:\n")
print(df_total_sales)
```

```python
# Q5.	Count the number of orders for each region.

# grouping by region
total_region = df.groupby('Region')

df_total_region = total_region.agg({'Order ID': 'count'}).sort_values(by = 'Order ID', ascending = False)

print("The total number of orders for each region:\n")
print(df_total_region)
```

```python
# Q6.	Calculate the average sales per quantity for the each category.

# finding out how many categories is there
all_category = df.groupby(['Category'])

all_sales = all_category.agg({'Sales': 'sum'})


all_quantity = all_category.agg({'Quantity': 'sum'})

avg_quantity_sold = round(all_sales['Sales']/all_quantity['Quantity'], 2)
avg_quantity_sold
```

```python
# Q7.	Identify the order with the highest profit.

highest_profit = df.sort_values(by = 'Total Profit', ascending = False).head(1)

print("The order with the highest profit is:\n")
highest_profit
```

```python
# Q8.	Count the number of orders that were shipped to the state of California.

df_cal = df.loc[df['State'] == 'California']['Order ID'].count()

print(f"The number of orders that were shipped to the state of California is: {df_cal}") 
```
